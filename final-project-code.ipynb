```python
#same process repeated from cell 1 except using a line equation
y=stock_price_df
x=new_df['eps_basic_net']
from sklearn.metrics import r2_score 
R_square = r2_score(x, y) 
print('Coefficient of Determination', R_square) 
m,b = np.polyfit(x, y, 1)
#instead of a polynomial with a power of two it is a linear regression and uses the info from polyfit to make an equation that graphs a line
y_1=m*x+b
plt.plot(x,y_1)
plt.xlabel('EPS')
plt.ylabel('Estimated Stock price')
plt.title('EPS in relation to stock price')
plt.scatter(new_df['eps_basic_net'],stock_price_df)
```
